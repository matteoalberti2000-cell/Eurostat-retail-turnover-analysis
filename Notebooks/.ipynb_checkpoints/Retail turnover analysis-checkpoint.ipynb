# Retail Turnover Analysis in Italy and Europe

This project analyzes the evolution of retail turnover indices in Italy using Eurostat data, with a comparison to France and Germany. The goal is to explore long-term trends, identify major downturns linked to economic crises, and measure how quickly each economy recovered.

**Tools used:** Python, Pandas, NumPy, Matplotlib, Eurostat data.

## Loading and cleaning the dataset

The original Eurostat file contains multiple variables compressed into a single metadata column. The first step is to separate these variables, clean the column names, and reshape the data into a long format suitable for analysis.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from IPython.display import display_html

df = pd.read_csv("../data/raw/estat_sts_trtu_a.tsv", sep="\t")

# Split metadata column
metadata = df.iloc[:, 0].str.split(",", expand=True)
metadata.columns = [
    "freq",
    "indicator",
    "sector",
    "adjustment",
    "unit",
    "country"
]

# Attach metadata to yearly columns
df = pd.concat([metadata, df.iloc[:, 1:]], axis=1)

# Clean column names
df.columns = df.columns.str.strip()

# Clean yearly values and convert them to numeric
for col in df.columns[6:]:
    df[col] = (
        df[col]
        .astype(str)
        .str.replace(r"[^\d.]", "", regex=True)
        .replace("", pd.NA)
    )
    df[col] = pd.to_numeric(df[col], errors="coerce")

# Reshape to long format
df_long = df.melt(
    id_vars=["freq", "indicator", "sector", "adjustment", "unit", "country"],
    var_name="year",
    value_name="turnover_index"
)

df_long["year"] = pd.to_numeric(df_long["year"], errors="coerce").astype("Int64")
```

## Choosing the best data set available
Eurostat provides the turnover index using different base years (2010=100, 2015=100 and 2021=100). Before proceeding with the analysis, the available series were compared on the same plot.

The three index series are normalized with different base years, so their absolute values are not directly comparable. This chart is only used to assess temporal coverage and continuity across methodologies.

```python
#Italy data coverage for the bases
italy_I10 = df_long[
    (df_long["country"] == "IT") &
    (df_long["indicator"] == "NETTUR") &
    (df_long["unit"] == "I10") &
    (df_long["adjustment"] == "CA") &
    (df_long["sector"] == "G")
]
italy_I10 = italy_I10.sort_values("year").reset_index(drop=True)


italy_I15 = df_long[
    (df_long["country"] == "IT") &
    (df_long["indicator"] == "NETTUR") &
    (df_long["unit"] == "I15") &
    (df_long["adjustment"] == "CA") &
    (df_long["sector"] == "G")
]
italy_I15 = italy_I15.sort_values("year").reset_index(drop=True)

italy_I21 = df_long[
    (df_long["country"] == "IT") &
    (df_long["indicator"] == "NETTUR") &
    (df_long["unit"] == "I21") &
    (df_long["adjustment"] == "CA") &
    (df_long["sector"] == "G")
]
italy_I21 = italy_I21.sort_values("year").reset_index(drop=True)

plt.figure(figsize=(11,6))
plt.plot(italy_I10["year"], italy_I10["turnover_index"], marker='o', color='g')
plt.plot(italy_I15["year"], italy_I15["turnover_index"], marker='o', color='r')
plt.plot(italy_I21["year"], italy_I21["turnover_index"], marker='o', color='b')
plt.grid(alpha=0.3)
plt.title("Italian data coverage for the series")
plt.xlabel("Year")
plt.ylabel("Turnover index")
plt.legend(['Base 2010', 'Base 2015','Base 2021'], loc=2)
plt.savefig("../Images/Italian data coverage for the series.png", dpi=300, bbox_inches="tight")
plt.show()
```

The I15 series will be used moving forward giving its coverage and continuity between 2000 and 2023.
The Italian retail turnover index shows a long-term upward trend, interrupted by three major downturns. These correspond to the global financial crisis, the European sovereign debt crisis, and the COVID-19 shock.

## Italy: long-term turnover analysis and interpretation

The first downturn occurred in 2009 and is associated with the global financial crisis triggered by the collapse of Lehman Brothers in September 2008. The resulting credit crunch and economic slowdown affected most developed economies, including Italy, leading to a visible reduction in retail activity.

The second contraction happened in 2012 and coincides with the European sovereign debt crisis. During the late stages of 2011 and throughout 2012, concerns regarding public debt sustainability across several Eurozone countries led to weaker economic growth, reduced consumer confidence and lower domestic demand in Italy.

The most pronounced decline is observed in 2020 and corresponds to the COVID-19 pandemic. The restrictions introduced to contain the spread of the virus significantly reduced economic activity and consumer spending, causing a sharp contraction in retail turnover.

Despite these shocks, the index shows a gradual recovery after each crisis, illustrating the resilience of the Italian retail sector over the long term.

##  Year-on-year growth

To quantify changes more precisely, I compute the annual percentage variation of the turnover index. This makes it easier to identify the strongest expansions and contractions in the series.

```python
italy_I15["growth_pct"] = (italy_I15["turnover_index"].pct_change()* 100)
italy_I15[["year", "turnover_index", "growth_pct"]]

plt.figure(figsize=(12,6))
plt.bar(
    italy_I15["year"],
    italy_I15["growth_pct"]
)

plt.title("Annual Growth Rate of Italian Retail Turnover")
plt.xlabel("Year")
plt.ylabel("Growth (%)")
plt.grid(alpha=0.3)
plt.savefig("../Images/Annual Growth Rate of Italian Retail Turnover.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
top_5_growth = italy_I15.nlargest(5, "growth_pct")[["year", "growth_pct"]]
worst_5_growth = italy_I15.nsmallest(5, "growth_pct")[["year", "growth_pct"]]
top_5_turnover = italy_I15.nlargest(5, "turnover_index")[["year", "turnover_index"]]
worst_5_turnover = italy_I15.nsmallest(5, "turnover_index")[["year", "turnover_index"]]
display_html(
    f"""
    <table>
    <tr>
        <td style="vertical-align:top; padding:20px;">
            <h3>Strongest Growth Years</h3>
            {top_5_growth.to_html(index=False)}
        </td>

        <td style="vertical-align:top; padding:20px;">
            <h3>Largest Contraction Years</h3>
            {worst_5_growth.to_html(index=False)}
        </td>
    </tr>

    <tr>
        <td style="vertical-align:top; padding:20px;">
            <h3>Highest Turnover Index Values</h3>
            {top_5_turnover.to_html(index=False)}
        </td>

        <td style="vertical-align:top; padding:20px;">
            <h3>Lowest Turnover Index Values</h3>
            {worst_5_turnover.to_html(index=False)}
        </td>
    </tr>
    </table>
    """,
    raw=True
)
```

## Recovery time

Beyond the size of each downturn, I also measure how long it takes the retail sector to return above pre-crisis levels. This provides a simple indicator of economic resilience.

```python
def recovery_time(df, crisis_year):
    crisis_row = df.loc[df["year"] == crisis_year, "turnover_index"]

    if crisis_row.empty:
        return None

    threshold = crisis_row.iloc[0]

    recovery = df[
        (df["year"] > crisis_year) &
        (df["turnover_index"] >= threshold)
    ]

    if recovery.empty:
        return None

    return int(recovery.iloc[0]["year"] - crisis_year)

recovery_2008_it = recovery_time(italy_I15, 2008)
recovery_2012_it = recovery_time(italy_I15, 2011)
recovery_2020_it = recovery_time(italy_I15, 2020)
print(f"Retail turnover got back to a pre-2008 crisis level in {recovery_2008_it} years.")
print(f"Retail turnover got back to a pre-2012 crisis level in {recovery_2012_it} years.")
print(f"Retail turnover got back to a pre-2020 crisis level in {recovery_2020_it} years.")
```

## France and Germany

The same methodology is applied to France and Germany in order to compare the evolution of the retail sector across three major European economies. This allows a direct comparison of crisis impact and recovery speed.

```python
france_I15 = df_long[
    (df_long["country"] == "FR") &
    (df_long["indicator"] == "NETTUR") &
    (df_long["sector"] == "G") &
    (df_long["adjustment"] == "CA") &
    (df_long["unit"] == "I15")
].copy().sort_values("year").reset_index(drop=True)

germany_I15 = df_long[
    (df_long["country"] == "DE") &
    (df_long["indicator"] == "NETTUR") &
    (df_long["sector"] == "G") &
    (df_long["adjustment"] == "CA") &
    (df_long["unit"] == "I15")
].copy().sort_values("year").reset_index(drop=True)

france_I15["growth_pct"] = france_I15["turnover_index"].pct_change() * 100
germany_I15["growth_pct"] = germany_I15["turnover_index"].pct_change() * 100

plt.figure(figsize=(11,6))
plt.plot(germany_I15["year"], germany_I15["turnover_index"], marker='o', color='y')
plt.plot(france_I15["year"], france_I15["turnover_index"], marker='o', color='r')
plt.grid(alpha=0.3)
plt.title("French and German retail turnover index")
plt.xlabel("Year")
plt.ylabel("Turnover index")
plt.legend(['Germany Base 2015','France Base 2015'], loc=2)
plt.savefig("../Images/French and German retail turnover index.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Cross-country resilience comparison

By comparing the magnitude of downturns and the time needed to recover, it is possible to assess which economy was hit the hardest and which one showed the strongest resilience over time.

To keep the comparison consistent, the analysis below uses the same Eurostat methodology for all three countries.

The comparison will focus on the 2008 and covid crises, where 

```python
comparison = italy_I15[["year", "growth_pct"]].rename(columns={"growth_pct": "Italy"})
comparison = comparison.merge(
    france_I15[["year", "growth_pct"]].rename(columns={"growth_pct": "France"}),
    on="year",
    how="inner"
)
comparison = comparison.merge(
    germany_I15[["year", "growth_pct"]].rename(columns={"growth_pct": "Germany"}),
    on="year",
    how="inner"
).sort_values("year")

crisis_2008 = comparison[
    (comparison["year"] >= 2007) &
    (comparison["year"] <= 2011)
].copy()
print(crisis_2008)
crisis_covid = comparison[
    (comparison["year"] >= 2018) &
    (comparison["year"] <= 2022)
].copy()
print(crisis_covid)
```

```python
x = np.arange(len(crisis_2008["year"]))
width = 0.25

fig, ax = plt.subplots(figsize=(11, 6))

bars_it = ax.bar(x - width, crisis_2008["Italy"], width, label="Italy")
bars_fr = ax.bar(x,         crisis_2008["France"], width, label="France")
bars_de = ax.bar(x + width, crisis_2008["Germany"], width, label="Germany")

ax.axhline(0, color="black", linewidth=1)
ax.set_xticks(x)
ax.set_xticklabels(crisis_2008["year"], rotation=45)
ax.set_ylabel("Growth (%)")
ax.set_title("Retail turnover growth during the global financial crisis")
ax.legend()
plt.tight_layout()
plt.savefig("../Images/Retail turnover growth during the global financial crisis.png", dpi=300, bbox_inches="tight")
plt.show()
```

```python
x = np.arange(len(crisis_covid["year"]))
width = 0.25
fig, ax = plt.subplots(figsize=(11, 6))
bars_it = ax.bar(x - width, crisis_covid["Italy"], width, label="Italy")
bars_fr = ax.bar(x,         crisis_covid["France"], width, label="France")
bars_de = ax.bar(x + width, crisis_covid["Germany"], width, label="Germany")

ax.axhline(0, color="black", linewidth=1)
ax.set_xticks(x)
ax.set_xticklabels(crisis_covid["year"], rotation=45)
ax.set_ylabel("Growth (%)")
ax.set_title("Retail turnover growth during the Covid-19 pandemic crisis")
ax.legend()
plt.tight_layout()
plt.savefig("../Images/Retail turnover growth during the Covid-19 pandemic crisis.png", dpi=300, bbox_inches="tight")
plt.show()
```

## Key findings

```python
def country_summary(df,recovery_2008,recovery_covid):
    avg_growth = df["growth_pct"].mean()
    volatility = df["growth_pct"].std()
    best_row = df.loc[df["growth_pct"].idxmax()]
    worst_row = df.loc[df["growth_pct"].idxmin()]
    return {
        "Average Growth (%)": round(avg_growth, 2),
        "Volatility": round(volatility, 2),
        "Best Year": int(best_row["year"]),
        "Best Growth (%)": round(best_row["growth_pct"], 2),
        "Worst Year": int(worst_row["year"]),
        "Worst Growth (%)": round(worst_row["growth_pct"], 2),
        "Recovery after 2008 (years)": recovery_2008,
        "Recovery after Covid (years)": recovery_covid
    }

summary_table = pd.DataFrame({
    "Italy": country_summary(
        italy_I15,
        recovery_time(italy_I15, 2008),
        recovery_time(italy_I15, 2019)
    ),
    "France": country_summary(
        france_I15,
        recovery_time(france_I15, 2008),
        recovery_time(france_I15, 2019)
    ),
    "Germany": country_summary(
        germany_I15,
        recovery_time(germany_I15, 2008),
        recovery_time(germany_I15, 2019)
    )
})

summary_table
```

The recovery metrics highlight important differences between economic shocks.
While all three countries eventually recovered from both crises, recovery times varied considerably. The 2008 financial crisis generally required longer recovery periods, whereas the Covid-19 shock was followed by a much faster rebound in retail activity.

Germany displayed the shortest recovery times overall, suggesting greater resilience of its retail sector.
France showed moderate growth and volatility, while Italy experienced longer recovery periods and higher sensitivity to economic shocks.
